```python
import sys  # Add this at the top of your file

cursor = db.cursor(dictionary=True)

print("Calculating mean GPA by program...")

# 1. First check how many empty GPA values exist
cursor.execute("""
    SELECT COUNT(*) as empty_count 
    FROM `student migration record` 
    WHERE GPA = '' OR GPA = ' ' OR TRIM(GPA) = ''
""")
empty_count = cursor.fetchone()['empty_count']
print(f"Found {empty_count} records with empty GPA")

if empty_count == 0:
    print("No empty GPA values found to update")
    sys.exit(0)  # Exit the script gracefully instead of using 'return'

# 2. Get mean GPA for each program (with student count)
cursor.execute("""
    SELECT ProgramOfStudy, 
           AVG(GPA) as mean_gpa,
           COUNT(*) as student_count
    FROM `student migration record`
    WHERE GPA != '' AND GPA != ' ' AND TRIM(GPA) != ''
    GROUP BY ProgramOfStudy
""")

program_means = {}
for row in cursor.fetchall():
    program = row['ProgramOfStudy']
    program_means[program] = {
        'mean': row['mean_gpa'],
        'count': row['student_count']
    }
    print(f"- {program}: Mean GPA {row['mean_gpa']:.2f} (based on {row['student_count']} students)")

# 3. Update empty GPAs with program means
total_updated = 0
for program, data in program_means.items():
    # First check if any empty GPAs exist for this program
    cursor.execute("""
        SELECT COUNT(*) as empty_count
        FROM `student migration record`
        WHERE ProgramOfStudy = %s 
        AND (GPA = '' OR GPA = ' ' OR TRIM(GPA) = '')
    """, (program,))
    program_empty_count = cursor.fetchone()['empty_count']
    
    if program_empty_count == 0:
        print(f"No empty GPAs found for {program}")
        continue
        
    # Perform the update
    cursor.execute("""
        UPDATE `student migration record`
        SET GPA = %s
        WHERE ProgramOfStudy = %s
        AND (GPA = '' OR GPA = ' ' OR TRIM(GPA) = '')
    """, (float(data['mean']), program))
    
    updated = cursor.rowcount
    total_updated += updated
    print(f"Updated {updated} {program} records with mean GPA {data['mean']:.2f}")

db.commit()
print(f"\nSuccessfully updated {total_updated} empty GPA values")
```

```python
invalid_phrases = [
    "not available",
    "not entered",
    "n/a",
    "na",
    "null",
    "missing",
    ""
]
column_name = "AdmissionTestScore"

# First verify the column exists
cursor.execute("SHOW COLUMNS FROM `student migration record` LIKE %s", (column_name,))
if not cursor.fetchone():
    print(f"Error: Column '{column_name}' does not exist in the table")
    exit()

# Create placeholders for the IN clause
placeholders = ", ".join(["%s"] * len(invalid_phrases))

# Execute the update
try:
    cursor.execute(f"""
        UPDATE `student migration record`
        SET `{column_name}` = 0
        WHERE LOWER(TRIM(`{column_name}`)) IN ({placeholders})
        OR `{column_name}` IS NULL
    """, tuple(invalid_phrases))
    
    updated_count = cursor.rowcount
    db.commit()
    print(f"Successfully updated {updated_count} records to 0")
    
except mysql.connector.Error as err:
    print(f"Database error occurred: {err}")
    db.rollback()
cursor.execute("""
    SELECT 
        Gender,
        AVG(`AdmissionTestScore`) as mean_score,
        COUNT(*) as student_count
    FROM `student migration record`
    WHERE `AdmissionTestScore` > 0  # Only consider valid scores
    GROUP BY Gender
""")

gender_means = {row['Gender']: row['mean_score'] for row in cursor.fetchall()}
print("Calculated mean scores by gender:")
for gender, mean in gender_means.items():
    print(f"- {gender}: {mean:.2f}")

# Step 3: Update zero scores with gender-specific means
for gender, mean_score in gender_means.items():
    cursor.execute("""
        UPDATE `student migration record`
        SET `AdmissionTestScore` = %s
        WHERE `AdmissionTestScore` = 0
        AND `Gender` = %s
    """, (mean_score, gender))
    print(f"Updated {cursor.rowcount} {gender} records with mean score {mean_score:.2f}")

db.commit()
```
